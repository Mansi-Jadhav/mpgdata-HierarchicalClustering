```python
df_w_dummies = pd.get_dummies(df.drop('name', axis=1))
```

```python
df_w_dummies
```

```python
from sklearn.preprocessing import MinMaxScaler # Transforms everything between 0 and 1
```

```python
scaler = MinMaxScaler()
```

```python
scaled_data = scaler.fit_transform(df_w_dummies)
```

```python
scaled_data
```

```python
scaled_df = pd.DataFrame(scaled_data, columns=df_w_dummies.columns)
```

```python
scaled_df
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(scaled_df)
```

```python
# Notice the origin columns - they're either 0 or 1 - so black and white
# model year - some specific years - still a range of number
# rest - continuous
```

```python
plt.figure(figsize=(15,8))
sns.clustermap(scaled_df)    # It has grouped columns and rows into features
# Notice dummy variables will be the last ones to be clustered - 0 and 1
```

```python
plt.figure(figsize=(15,8))
sns.clustermap(scaled_df, row_cluster=False)   # Cluster based on features
# Similar to heatmap based on correlation
```

```python
plt.figure(figsize=(15,8))
sns.clustermap(scaled_df, col_cluster=False)   # Cluster based on rows - This is what we're looking for

# We won't be able to do this for large datasets
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=4)  # Let's choose 4 to get those 4 clusters from clustermap in level 3
```

```python
cluster_labels = model.fit_predict(scaled_df)
```

```python
cluster_labels   # 0,1,2,3
```

```python
plt.figure(figsize=(12,6), dpi=200)
sns.scatterplot(data=df, x='mpg', y='weight', hue=cluster_labels, palette='viridis')

# As weight increases, mpg decreases
```

```python
plt.figure(figsize=(12,6), dpi=200)
sns.scatterplot(data=df, x='mpg', y='horsepower', hue=cluster_labels, palette='viridis')
```

```python
# Max distance possible?
# Square root of number of features
# D^2 = (x11-x12)^2 + ...... (x(n-1)(p-1) - xnp)^2   # Euclidean (Similar to pythagorus theorem)
# x11 - x12 -> Max value will be 1 - scaling
# So 1+1+..
```

```python
len(scaled_df.columns)
```

```python
np.sqrt(10)    # Max distance possible between 2 points
```

```python
# Distance threshold - Linkage distance threshold above which clusters will not be merged
# If it's 0, none of the points will be merged - (even if distance between points in 0 units, they will still not be merged)
```

```python
model = AgglomerativeClustering(n_clusters=None, distance_threshold=0)
```

```python
cluster_labels = model.fit_predict(scaled_df)
```

```python
cluster_labels   # Unique cluster for each point
```

```python
from scipy.cluster import hierarchy
from scipy.cluster.hierarchy import dendrogram
```

```python
linkage_matrix = hierarchy.linkage(model.children_)
```

```python
linkage_matrix     # point1, point2, distance betn points, number of points
```

```python
plt.figure(figsize=(20,10), dpi=200)
dendro = dendrogram(linkage_matrix)

# y axis length is the distance betn points
# 75 is the highest one - last in the linkage matrix
```

```python
# We can use truncate mode to truncate at some point - lastp or level  (documentation)
```

```python
plt.figure(figsize=(20,10), dpi=200)
dendro = dendrogram(linkage_matrix, truncate_mode='lastp', p=10)
# It chooses last 10 points for hierarchy
```

```python
plt.figure(figsize=(20,10), dpi=200)
dendro = dendrogram(linkage_matrix, truncate_mode='level', p=10)
```
